```python
import os
import requests
import json
import time

SERVICE_KEY = "48RysNl1ik0Yayfh1ih7lOwKPbSQmus+nhN76dcuw/5LSr4uXnBrSZEPYq5QsUWf/37wyPGCbmOawE5X2mMeBw=="
BASE_URL = "https://apis.data.go.kr/1611000/undergroundsafetyinfo"

def fetch_all_pages(endpoint: str, filename_prefix: str, params: dict, key_path: list, max_pages=100):
    all_items = []
    page = 1

    while page <= max_pages:
        params["pageNo"] = page
        params["numOfRows"] = 1000
        params["type"] = "json"

        try:
            url = f"{BASE_URL}/{endpoint}"
            response = requests.get(url, params={**params, "serviceKey": SERVICE_KEY})
            response.raise_for_status()
            data = response.json()
        except Exception as e:
            print(f"âŒ ìš”ì²­ ì‹¤íŒ¨ (í˜ì´ì§€ {page}): {e}")
            break

        items = data
        for key in key_path:
            items = items.get(key, [])

        if not items:
            break

        all_items.extend(items)
        if len(items) < params["numOfRows"]:
            break

        page += 1
        time.sleep(0.1)

    # ì €ì¥
    os.makedirs("Database/openapi", exist_ok=True)
    output_path = f"Database/openapi/{filename_prefix}_ì „ì²´.json"
    with open(output_path, "w", encoding="utf-8") as f:
        json.dump(all_items, f, ensure_ascii=False, indent=2)
        print(f"âœ… ì „ì²´ ì €ì¥ ì™„ë£Œ: {output_path}")

    return all_items
```

```python
import math
import requests

def get_total_pages(endpoint: str, params: dict, rows_per_page: int = 1000) -> int:
    BASE_URL = "https://apis.data.go.kr/1611000/undergroundsafetyinfo"
    SERVICE_KEY = "48RysNl1ik0Yayfh1ih7lOwKPbSQmus+nhN76dcuw/5LSr4uXnBrSZEPYq5QsUWf/37wyPGCbmOawE5X2mMeBw=="

    full_params = {
        "serviceKey": SERVICE_KEY,
        "pageNo": 1,
        "numOfRows": 1,
        "type": "json"
    }
    full_params.update(params)

    try:
        response = requests.get(f"{BASE_URL}/{endpoint}", params=full_params)
        response.raise_for_status()
        data = response.json()
        total_count = int(data["response"]["totalCount"])
        total_pages = math.ceil(total_count / rows_per_page)
        print(f"ğŸ“Š ì „ì²´ ë°ì´í„° ìˆ˜: {total_count}ê°œ â†’ ì´ {total_pages} í˜ì´ì§€ ì˜ˆìƒ (í˜ì´ì§€ë‹¹ {rows_per_page}ê°œ)")
        return total_pages
    except Exception as e:
        print(f"âŒ ì „ì²´ í˜ì´ì§€ ìˆ˜ ê³„ì‚° ì˜¤ë¥˜: {e}")
        return -1
```
